```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../files/labels.csv")
```

```python
label = {k: v for (k, v) in zip(set(df.iloc[:1000,2]),np.full(21,0,dtype=int))}
```

```python
for i, el in df.iloc[:1000].iterrows():
    label[el[2]] = label[el[2]] + 1
```

```python
# Data to plot
labels = []
sizes = []
for x, y in label.items():
    labels.append(x)
    sizes.append(y)

# Plot
plt.figure(figsize=(9,9))
patches, texts, autotexts = plt.pie(sizes,
        textprops={'size': 'medium'},
        colors=['#eae4e9', '#fff1e6', '#fde2e4',
                '#fad2e1', '#e2ece9', '#bee1e6',
                '#cddafd', '#dfe7fd','#bde0fe'],
        labels=labels, 
        autopct='%1.1f%%')
for text in texts:
    text.set_color('#555555')
for autotext in autotexts:
    autotext.set_color('#555555')
plt.axis('equal')
plt.savefig('labels.png')
plt.show()
```
